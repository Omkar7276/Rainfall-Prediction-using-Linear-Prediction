# Rainfall Prediction using Linear Regression

```python
# importing the libraries

import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

# Data Cleaning

```python
#drop the unneccessary columns

data =  data.drop(['Events','Date','SeaLevelPressureLowInches'], axis=1)
```

```python
#replacing the value 'T' and '-'' to the zero
data = data.replace('T', 0.0)
data = data.replace('-',0.0)
```

```python
data.to_csv('austin_final.csv')
```

# Data Analysis

```python
# importing the libraries for data analysis
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('austin_final.csv')
data
```

```python
X = data.drop(['PrecipitationSumInches'], axis=1)
```

```python
Y = data['PrecipitationSumInches']
```

```python
Y
```

```python
# now we will reshape it into the 2D vector format

Y = Y.values.reshape(-1,1)
```

```python
# CONSIDER THE RANDOM DAY OF THE DATASETS


day_index = 798

days = [i for i in range(Y.size)]
```

# Initializing the Linear Regression Classifier

```python
clf = LinearRegression()


# train the classifier

clf.fit(X,Y)
```

```python
# plot a graph

print('Precipitation trend graph')
plt.scatter(days, Y, color = 'r')
plt.scatter(days[day_index], Y[day_index], color = 'b')
plt.title('Precipitation Level')
plt.xlabel('Days')
plt.ylabel('Precipitation in inches')
plt.show()
x_vis = X.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
print('The Precipitation vs attributes ternd graph : ')
for i in range (x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color = 'r')
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color = 'b')
    plt.title(x_vis.columns[i])
plt.show()
```
